```python
#multiclass classification
#softmax function 
#softmax = e(x) / sum(e^(x))
#creates multiple functions for each class. Input into each and the largest percentage indicates the class it would be for the output classification.
```

```python
import torch
import torch.nn as nn
```

```python
class LinearClassifier(torch.nn.Module):
    def __init__(self, input_dim, n_classes):
        super().__init__()
        self.fc = torch.nn.Linear(input_dim, n_classes)

    def forward(self, x):
        return nn.functional.softmax(self.fc(x), dim=-1) #SOFTMAX is the key point
    
model = LinearClassifier(10, 4)
print(model)
print(model.fc.weight)
print(model.fc.bias)

#x= torch.zeros(10)
x=torch.ones(10)
model(x)
```

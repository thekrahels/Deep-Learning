```python
#sigma x function is the sigmoid function
#sigmoid function is used to convert the output of a linear model into a probability between 0 and 1
#sigmoid function is defined as 1 / (1 + e^(-x))
#where x is the output of the linear model (the weighted sum of the inputs)
#binary linear function is sigmoid(W * x + b)
#So take linear and mutilply it by sigmoid to get the output of the model in the range of 0 to 1 (binary)
```

```python
import torch
import torch.nn as nn
```

```python
class LinearClassifier(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.fc = torch.nn.Linear(input_dim, output_dim)

    def forward(self, x):
        return nn.functional.sigmoid(self.fc(x)) #never use this this!!!!! sigmoid should not be used directly
    
model = LinearClassifier(10, 1)
print(model)
print(model.fc.weight)
print(model.fc.bias)

#x= torch.zeros(10)
x=torch.ones(10)
model(x)
```

```python
import torch
```

```python
#4 layer model

class MyModel(torch.nn.Module):
    def __init__(self, layer_size=[512, 512, 512]): #first 3 layers
        super().__init__()
        layers = []
        layers.append(torch.nn.Flatten())
        c = 128 * 128 * 3
        for s in layer_size:
            layers.append(torch.nn.Linear(c, s))
            layers.append(torch.nn.ReLU())
            c = s
        layers.append(torch.nn.Linear(c, 102)) #final layer
        self.model = torch.nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)


net = MyModel()
x = torch.randn(10, 3, 128, 128) #random inputs 
net(x)
```

```python
x = torch.randn(10, 3, 128, 128)
net0 = MyModel([]) #model with no hidden layers

#expand the layers
net1 = MyModel([512]) #2 layer model with 1 hidden layer of 512 neurons

#print(net1(x))
#can continue this trend and slowly the outputs will get smaller and smaller 
#this is why we need a normalization layer to keep the outputs in a reasonable range

#can also add more layers and make the model deeper
net2 = MyModel([512, 512]) #3 layer model with 2 hidden layers of 512 neurons

#print the norm to show it gets smaller and smaller as we add more layers
print(f"{net0(x).norm()=}")
print(f"{net1(x).norm()=}")
print(f"{net2(x).norm()=}")

#add more layers using a for loop
for n in range(10):
    netn = MyModel([512] * n) #model with n hidden layers of 512 neurons
    print(f"{netn(x).norm()=}")

    #shows quickly how they continue to go towards zero as we add more layers, this is why we need normalization layers to keep the outputs in a reasonable range
    #if the output becomes zero then when that goes into the next layer then it becomes a model that does nothing for us because input will be zero and output will be zero
    #it will stop to go lower because the bias which is the layers.append(torch.nn.Linear(c, s)) needs to be layers.append(torch.nn.Linear(c, s, bias=False)) to remove the bias and then it will continue to go towards zero as we add more layers
    # additionally put bias as false for the final layer of the model
    
```

```python
#add a batch normalization layer 
# After each linear layer   

class MyModelBN(torch.nn.Module):
    def __init__(self, layer_size=[512, 512, 512]):
        super().__init__()
        layers = []
        layers.append(torch.nn.Flatten())
        c = 128 * 128 * 3
        for s in layer_size:
            layers.append(torch.nn.Linear(c, s))
            layers.append(torch.nn.BatchNorm1d(s)) ################this is a batch normalization layer
            layers.append(torch.nn.ReLU())
            c = s
        layers.append(torch.nn.Linear(c, 102))
        self.model = torch.nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)


x = torch.randn(10, 3, 128, 128)
for n in range(10):
    netn = MyModelBN([512] * n)
    print(f"{n=}  {netn(x).norm()=}")
```

```python
#after the normalization, we disable the bias
#normalization will just remove the bias so:
# either we have + a bias and then normalization removes the bias
# or we have no bias in the beginning
#note that if you place the normalization before the ReLu then this works
#if you put it before the linear layer then the linear layer needs 


#add a batch normalization layer 
# After each linear layer   

class MyModelBN(torch.nn.Module):
    def __init__(self, layer_size=[512, 512, 512]):
        super().__init__()
        layers = []
        layers.append(torch.nn.Flatten())
        c = 128 * 128 * 3
        for s in layer_size:
            layers.append(torch.nn.Linear(c, s, bias=False)) ##########this is the bias removed
            layers.append(torch.nn.BatchNorm1d(s)) ################this is a batch normalization layer
            layers.append(torch.nn.ReLU())
            c = s
        layers.append(torch.nn.Linear(c, 102, bias=False))  ##########this is the bias removed
        self.model = torch.nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)


x = torch.randn(10, 3, 128, 128)
for n in range(10):
    netn = MyModelBN([512] * n)
    print(f"{n=}  {netn(x).norm()=}")
```

```python
class MyModelLN(torch.nn.Module):
    def __init__(self, layer_size=[512, 512, 512]):
        super().__init__()
        layers = []
        layers.append(torch.nn.Flatten())
        c = 128 * 128 * 3
        for s in layer_size:
            layers.append(torch.nn.Linear(c, s))
            layers.append(torch.nn.LayerNorm(s, bias=False)) #############this is the difference between batch normalization and layer normalization
            layers.append(torch.nn.ReLU())
            c = s
        layers.append(torch.nn.Linear(c, 102, bias=False))
        self.model = torch.nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)


x = torch.randn(10, 3, 128, 128)
for n in range(30):
    netn = MyModelLN([512] * n)
    print(f"{n=}  {netn(x).norm()=}")

# goal is for the normalization to prevent the output from going to zero
#19 goes to 13 because of training then stays in that range because of the normalization
```

```python

#place the normalization before the linear layer
#norm needs to be bias false
#linear needs bias on
# bias should always be there before the ReLu


class MyModelLNBeforeLinear(torch.nn.Module):
    def __init__(self, layer_size=[512, 512, 512]):
        super().__init__()
        layers = []
        layers.append(torch.nn.Flatten(start_dim=1))#added this
        c = 128 * 128 * 3
        for s in layer_size:
            layers.append(torch.nn.LayerNorm(s, bias=False, eps = 1e-5, elementwise_affine=True)) #############this is the difference and added the eps and elementwise_affine
            layers.append(torch.nn.Linear(c, s, bias = True)) 
            ##layers.append(torch.nn.LayerNorm(s, bias=False)) #############this is the difference
            layers.append(torch.nn.ReLU())
            c = s
        layers.append(torch.nn.Linear(c, 102, bias=False))
        self.model = torch.nn.Sequential(*layers)

    def forward(self, xBeforeLinear):
        x = xBeforeLinear.view(xBeforeLinear.size(0), -1)
        return self.model(x)


xBeforeLinear = torch.randn(10, 3, 128, 128)
for n in range(30):
    netnBeforeLinear = MyModelLNBeforeLinear([512] * n)
    print(f"{n=}  {netnBeforeLinear(xBeforeLinear).norm()=}")
```

```python
import torch
```

```python
model = torch.nn.Linear(10, 1)
```

```python
x = torch.randn(20, 10)
y = torch.randn(20, 1)
print(f'{x=}{y=}')
```

```python
pred_y = model(x)
print(pred_y)
```

```python
#compare y and pred_y
loss = torch.nn.functional.mse_loss(pred_y, y)
print(loss)
#loss = loss_fn(pred_y, y)
```

```python
#actual breakdown of the calculation from above function
def mse_loss(pred_y, y):
    return torch.mean((pred_y - y) ** 2)
mse_loss(pred_y, y)
```

```python
#binary loss

y = (torch.randn(20, 1) > 0).float()
print(y)
```

```python
#adds sigmoid for us - do not add it ourselves
#.7 is essentially random so 50/50 chance gets .7
# less than .7 is better than random, more than .7 is worse than random

loss = torch.nn.functional.binary_cross_entropy_with_logits(pred_y, y)
print(loss)
```

```python
import torch

# example of l2 loss function for linear regression


def l2_loss(y_true, y_pred):
    return torch.sum((y_true - y_pred) ** 2)


linear = torch.nn.Linear(4, 1)
x = torch.rand((1, 4))

y_true = torch.as_tensor([1.0], dtype=torch.float32)
y_pred = linear(x)

print(l2_loss(y_true, y_pred))
```

```python
import torch

# example of binary cross entropy loss function for binary classification

linear = torch.nn.Linear(4, 1)
x = torch.rand((1, 4))

y_true = torch.as_tensor([[1]], dtype=torch.float32)

y_pred = torch.sigmoid(linear(x))

loss_func = torch.nn.BCELoss()

print(loss_func(y_pred, y_true))
```

```python
import torch

# example of multi-class cross entropy loss function for multi-class classification

num_class = 3

linear = torch.nn.Linear(4, num_class)
x = torch.rand((1, 4))

y_true = torch.as_tensor([2], dtype=torch.int64)

y_pred = linear(x)

loss_func = torch.nn.CrossEntropyLoss()

print(loss_func(y_pred, y_true))
```
